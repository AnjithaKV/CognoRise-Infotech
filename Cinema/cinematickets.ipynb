# **CINEMA TICKETS**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('/content/cinemaTicket_Ref.csv')
df
```

Dropping rows with missing values

```python
df.dropna(inplace=True)
df.isna().sum()
```

Droping columns 'month' and 'day' as they are unnecessary

```python
df.drop(['month','day'],inplace=True,axis=1)
df
```

Finding Correlation and showing it in heatmap

```python
df.corr()
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(df.corr(),annot=True)
plt.show()
```

As the correlation between ticket_use and tickets_sold, one of them is dropped

```python
df.drop(['ticket_use'],inplace=True,axis=1)
df
```

```python
df['date']=df['date'].str.replace('-','')
df['date']=df['date'].astype(int)
```

Splitting data into x and y

```python
x=df.drop(['total_sales'],axis=1)
x
```

```python
y=df['total_sales']
y
```

Regression plot of highly correlated features with target

```python
sns.regplot(x=df['tickets_sold'],y=y,color='red')
plt.title('Tickets_sold - Total_sales Regression Plot',color='blue')
```

```python
sns.regplot(x=df['show_time'],y=y,color='green')
plt.title('Show_time - Total_sales Regression Plot',color='red')
```

Regression plot of least correlated features with target

```python
sns.regplot(x=df['cinema_code'],y=y,color='pink')
plt.title('Cinema_code - Total_sales Regression Plot',color='green')
```

FEATURE SELECTION MUTUAL INFORMATION GAIN

```python
from sklearn.feature_selection import mutual_info_classif
mutual_info=mutual_info_classif(x,y,random_state=42)
mutual_info
```

```python
mutual_info=pd.Series(mutual_info)
mutual_info.index=x.columns
mutual_info.sort_values(ascending=False)
```

```python
mutual_info.sort_values(ascending=False).plot.bar()
```

We can check the relationship using regression plot

```python
sns.regplot(x=df['tickets_out'],y=y,color='black')
plt.title('Tickets_out - Ticket_sales Regression Plot')
```

From the regression plot we can we that both these features are not linearly dependent, so the feature can be dropped

```python
df.drop(['tickets_out'],inplace=True,axis=1)
```

Splitting imto training and testing data

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

MODEL CREATION

LINEAR REGRESSION

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

PERFORMANCE EVALUATION IN LINEAR REGRESSION

```python
from sklearn.metrics import mean_absolute_percentage_error,r2_score,mean_absolute_error,mean_squared_error
print('MAPE is',mean_absolute_percentage_error(y_test,y_pred))
print('R2 SCORE is',r2_score(y_test,y_pred))
print('MAE is',mean_absolute_error(y_test,y_pred))
print('MSE is',mean_squared_error(y_test,y_pred))
```

DECISION TREE

```python
from sklearn.tree import DecisionTreeRegressor
dt=DecisionTreeRegressor(criterion='absolute_error')
dt.fit(x_train,y_train)
y_pred=dt.predict(x_test)
y_pred
```

PERFORMANCE EVALUATION IN DECISION TREE

```python
from sklearn.metrics import mean_absolute_percentage_error,r2_score,mean_absolute_error,mean_squared_error
print('MAPE is',mean_absolute_percentage_error(y_test,y_pred))
print('R2 SCORE is',r2_score(y_test,y_pred))
print('MAE is',mean_absolute_error(y_test,y_pred))
print('MSE is',mean_squared_error(y_test,y_pred))
```

DECISION TREE PERFORMANCE IS BEST
